```python
f = open("C:/Users/94883/Desktop/Workshop/part-r-00000", 'r')
lines = f.readlines()
```

```python
word = []
file_name = []
value = []
for line in lines:
    word.append(line.split()[0])
    file_name.append(line.split()[1])
    value.append(line.split()[2])
```

```python
import pandas as pd
df = pd.DataFrame({'word':word,'file_name':file_name,'value':value})
```

```python
a =df.groupby('word',sort=False).apply(lambda x:x.sort_values('value',ascending=False)).reset_index(drop=True)
```

```python
a = a[['word','file_name']]
a.head()
```

```python
c = pd.merge(df1,a,on=['word','file_name'])
```

```python
c.head()
```

```python
c['new'] =c['file_name']+" "+c['value']
```

```python
c.head()
```

```python
c.drop(columns=['file_name','value'])
```

```python
b= c.groupby('word').agg({'new': ','.join }).reset_index()
```

```python
keys = list(b['word'])
values = list(b['new'].str.split(","))
```

```python
my_dict = dict(zip(keys,values))
```

```python
fw = open("big_index_dictionary_tfidf_2.txt",'wb')
fw.write(str(my_dict).encode('gbk'))      
fw.close()
```
